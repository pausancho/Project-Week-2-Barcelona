## Barcelona citizens with ages +60

```python
import pandas as pd
```

```python
url_padro = "https://opendata-ajuntament.barcelona.cat/data/dataset/9e93c1db-118a-4520-99e6-b3b955c33758/resource/61d91b31-9ba9-45ca-9cc3-47bf99555b4b/download/2020_padro_edat_any_a_any_per_sexe.csv"
```

```python
padro = pd.read_csv(url_padro)
```

```python
padro
```

```python
# Adding a key column to the data set
```

```python
padro["key"] = range(1, len(padro) + 1)
```

```python
padro.head()
```

```python
padro.columns
```

```python
padro_columns = ['any', 'codi_districte', 'nom_districte', 'codi_barri', 'nom_barri',
       'sexe', 'edat', 'nombre', "key"]
```

```python
padro.columns = padro_columns
```

```python
padro.head()
```

```python
years = pd.DataFrame(padro["edat"].str.split(" ", 1).tolist(),columns=["edats", "ref"])
```

```python
years["key"] = range(1, len(years) + 1)
```

```python
years
```

```python
# Mergin both data sets with the new column for the ages
```

```python
padro_merge = pd.merge(padro, years, on='key')
```

```python
padro_merge.head()
```

```python
# Dropping unecessary columns
```

```python
padro_merge.drop(["ref"], axis=1, inplace=True)
```

```python
padro_merge.drop(["edat"], axis=1, inplace=True)
```

```python
padro_merge.head()
```

```python
# Converting the object edats to int
```

```python
padro_merge['edats'] = padro_merge.edats.astype(int)
```

#### Total Population by neightborhood

```python
new_barri = barri.reset_index()
```

### Total Population by district

### Total population by age

```python
mayor_60 = padro_merge[padro_merge["edats"] > 60]
```

```python
total_60 = mayor_60.groupby("nom_districte")["nombre"].sum().sort_values()
```

```python
total_60df = pd.DataFrame(total_60).reset_index()
```

```python
total_60df
```

```python
percentage_60 = pd.merge(total_districte, total_60df, on='nom_districte')
```

```python
percentage_60 = percentage_60.rename(columns = {"nombre_x":"poblacio_total",
                                                "nombre_y":"poblacio_+60",
                                                })
```

```python
percentage_60["%_poblacion_+60"] = ((percentage_60["poblacio_+60"]/percentage_60["poblacio_total"])*100)
```

```python
#percentage_60.drop(["percentage"], axis=1, inplace=True)
```

```python
percentage_60.loc['Total']= percentage_60[["poblacio_total", "poblacio_+60"]].sum()
```

```python
percentage_60["+60_%_total"] = (percentage_60["poblacio_+60"]/sum(percentage_60.drop("Total")["poblacio_total"]))*100
```

```python
percentage_60.fillna("", inplace=True)
```

### Population Density

```python
pop_url = "https://opendata-ajuntament.barcelona.cat/data/dataset/b7ba32eb-806e-4c9c-b0b1-9bab387fe501/resource/08c88565-5b50-47d1-8234-6d3dd2da4c58/download/2019_densitat.csv"
```

```python
density = pd.read_csv(pop_url)
```

```python
density.head()
```

```python
lower_names = [i.lower() for i in density.columns]
```

```python
density.columns = lower_names
```

```python
district_density =  density.groupby("nom_districte").sum().reset_index()
```

```python
district_density
```

```python
real_pop = new_districte[["nom_districte", "nombre"]]
```

```python
real_density = pd.merge(district_density, real_pop, on='nom_districte')
```

```python
real_density
```

```python
pop_2020 = real_density.drop(["població"], axis=1)
```

```python
pop_2020['Km2'] = pop_2020["superfície (ha)"] * 0.01
```

```python
pop_2020["density"] = (pop_2020["nombre"]/pop_2020["Km2"]).round()
```

```python
pop_2020.loc['Total']= pop_2020[["superfície (ha)", "superfície residencial (ha)", "nombre", "Km2", "density"]].sum()
```

```python
pop_2020.drop(["densitat neta (hab/ha)"], axis=1, inplace=True)
```

```python
pop_2020.drop(["densitat (hab/ha)"], axis=1, inplace=True)
```

```python
pop_2020.drop(["codi_districte", "codi_barri", "any",], axis=1, inplace=True)
```

```python
pop_2020.fillna("", inplace=True)
```

```python
pop_2020
```

```python
pop_2020["poblacio_+60"] = percentage_60["poblacio_+60"]
```

```python
pop_2020
```

```python
pop_2020["%_poblacion_+60"] = percentage_60["%_poblacion_+60"]
```

```python
pop_2020["+60_%_total"] = percentage_60["+60_%_total"]
```

```python
pop_2020
```

```python
pop_2020.sort_values(["poblacio_+60"], ascending=False).drop("Total").reset_index()
```

```python
pop_2020["%_total_hab"] = (pop_2020["nombre"]/sum(pop_2020.drop("Total")["nombre"]))*100
```

```python
pop_2020
```

```python
pop_2020["%_total_density"] = (pop_2020["density"]/sum(pop_2020.drop("Total")["density"]))*100
```
