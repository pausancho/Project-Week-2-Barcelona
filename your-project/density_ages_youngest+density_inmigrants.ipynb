## Barcelona citizens

```python
import pandas as pd
```

```python
url_citiziens = "https://opendata-ajuntament.barcelona.cat/data/dataset/9e93c1db-118a-4520-99e6-b3b955c33758/resource/61d91b31-9ba9-45ca-9cc3-47bf99555b4b/download/2020_padro_edat_any_a_any_per_sexe.csv"
```

```python
citiziens_bcn_pd = pd.read_csv(url_citiziens)
```

```python
# Adding a key column to the data set
```

```python
citiziens_bcn_pd["index"] = range(1, len(citiziens_bcn_pd) + 1)
```

```python
citiziens_bcn_pd.head()
```

```python
citiziens_bcn_pd.columns
```

```python
citiziens_bcn_pd_columns = ['any', 'codi_districte', 'nom_districte', 'codi_barri', 'nom_barri',
       'sexe', 'edat', 'num_habitants', "index"]
```

```python
citiziens_bcn_pd.columns = citiziens_bcn_pd_columns
```

```python
citiziens_bcn_pd.head()
```

```python
years = pd.DataFrame(citiziens_bcn_pd["edat"].str.split(" ", 1).tolist(),columns=["edats", "ref"])
```

```python
years["index"] = range(1, len(years) + 1)
```

```python
# Mergin both data sets with the new column for the ages
```

```python
citiziens_bcn_pd_merge = pd.merge(citiziens_bcn_pd, years, on='index')
```

```python
citiziens_bcn_pd_merge.head()
```

```python
# Dropping unecessary columns
```

```python
citiziens_bcn_pd_merge.drop(["ref"], axis=1, inplace=True)
```

```python
citiziens_bcn_pd_merge.head()
```

```python
# Converting the object edats to int
```

```python
citiziens_bcn_pd_merge['edats'] = citiziens_bcn_pd_merge.edats.astype(int)
```

#### Total Population by neightborhood

```python
barri = citiziens_bcn_pd_merge.groupby("nom_barri").sum()
```

```python
new_barri = barri.reset_index()
```

```python
new_barri.head()
```

```python
#Llista de barris amb nombre d´habitans

new_barri[["nom_barri", "num_habitants"]]
```

### Total Population by district 

```python
districte = citiziens_bcn_pd_merge.groupby(["nom_districte"]).sum()
```

```python
new_districte = districte.reset_index()
```

```python
#Llista de districtes amb nombre d´habitants

new_districte[["nom_districte","num_habitants"]]
```

### Total population by age between 18 and 31

```python
citiziens_bcn_pd_merge.head(5)
```

```python
youngest = citiziens_bcn_pd_merge[(citiziens_bcn_pd_merge["edats"] >= 18) & (citiziens_bcn_pd_merge["edats"] <= 32) ]
```

### Número de joves per districte

```python
frec_youngest = youngest.groupby(['nom_districte'])['num_habitants'].sum().sort_values(ascending=False)
youngest_df = pd.DataFrame(data = frec_youngest).reset_index()
#youngest_df_frec.head()
#Merge de la taula total habitants x districte versus total habitant entre 18 y 32
percentatge_18_32= pd.merge(new_districte[[ 'nom_districte',"num_habitants",]], youngest_df, on='nom_districte')
#percentatge_18_32.drop('nom_barri_x',axis=1, inplace=True)
```

```python
percentatge_18_32 = percentatge_18_32.rename(columns={'num_habitants_x':'poblacio_total', 'num_habitants_y':'poblacio_[18-32]' })

#% del num de joves en un barri
percentatge_18_32['%_barris[18-32]'] = ((percentatge_18_32["poblacio_[18-32]"]/percentatge_18_32["poblacio_total"])*100).round()
#% del num de jove sobre el total de poblacio en el mateix barri
percentatge_18_32['%_barris[Joves/Total]'] = (percentatge_18_32["poblacio_[18-32]"]/sum(percentatge_18_32["poblacio_total"]))*100

percentatge_18_32
```

```python
#Sumar total i treure els NA
percentatge_18_32.loc['Total'] = percentatge_18_32[['poblacio_total','poblacio_[18-32]','%_barris[18-32]', '%_barris[Joves/Total]' ]].sum()
percentatge_18_32.fillna('',inplace=True)
percentatge_18_32.head()
```

### Density inmigrants

```python
url_inmigrants = 'https://opendata-ajuntament.barcelona.cat/data/dataset/653e65cc-9bda-4d89-9701-033998379066/resource/816f6f82-0598-4ff4-a4b8-1be4c898fd24/download/2019_immigrants_nacionalitats.csv'
```

```python
inmigrants_bcn_pd = pd.read_csv(url_inmigrants)
sum(inmigrants_bcn_pd['Nombre'])
```

```python
inmigrants_estrangers = inmigrants_bcn_pd[inmigrants_bcn_pd['Nacionalitats'] != 'Espanyola'].reset_index()
frec_df = inmigrants_estrangers.groupby(['Nom_Districte', 'Nom_Barri'])['Nombre'].sum().sort_values(ascending=False)
inmigrants_df_frec = pd.DataFrame(data = frec_df).reset_index()
sum(inmigrants_df_frec['Nombre'])
```

```python

inmigrants_df_frec['%_barris'] = round((inmigrants_df_frec["Nombre"]/sum(inmigrants_bcn_pd['Nombre'])*100),2)

inmigrants_df_frec.head()
```
